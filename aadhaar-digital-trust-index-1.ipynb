```python
enrol.drop('age_0_5', axis=1, inplace=True)
```

```python
df_bio = bio.groupby(['state', 'district' , 'pincode']).sum().reset_index()
```

```python
final_bio = df_bio.groupby('district').agg({
    'pincode' : 'first',
    'state': 'first',      # Keep the first state name found for that district
    'bio_age_5_17': 'sum' , # Sum the counts
    'state': 'first',      # Keep the first state name found for that district
    'bio_age_17_': 'sum'  # Sum the counts
}).reset_index()
```

```python
df_demo =  demo.groupby([  'state' , 'district' , 'pincode' ]).sum().reset_index()
```

```python
final_demo = df_demo.groupby('district').agg({
    'pincode' : 'first',
    'state': 'first',      # Keep the first state name found for that district
    'demo_age_5_17': 'sum' , # Sum the counts
    'state': 'first',      # Keep the first state name found for that district
    'demo_age_17_': 'sum'  # Sum the counts
}).reset_index()
```

```python
df_enrol =  enrol.groupby([  'state' , 'district' , 'pincode' ]).sum().reset_index()
```

```python
final_enrol = df_enrol.groupby('district').agg({
    'pincode' : 'first',
    'state': 'first',      # Keep the first state name found for that district
    'age_5_17': 'sum' , # Sum the counts
    'state': 'first',      # Keep the first state name found for that district
    'age_18_greater': 'sum'  # Sum the counts
}).reset_index()
```

```python
df = pd.merge(
    final_demo, 
    final_bio, 
    on=['state', 'district', 'pincode'], 
    how='inner'
)
```

```python
df.head()
```

```python
bio_age_sum = ['bio_age_5_17' , 'bio_age_17_']
df['bio_total']= df[bio_age_sum].sum(axis=1)
```

```python
demo_age_sum = ['demo_age_5_17','demo_age_17_']
df['demo_total']= df[demo_age_sum].sum(axis=1)
```

```python
df.head()
```

```python
#Digital Trust Index (DTI) 
df['DTI'] = df['demo_total'] / df['bio_total']
```

```python
df.head()
```

```python
df = df.sort_values(by='DTI')
```

```python
min_val = df['DTI'].min()
max_val = df['DTI'].max()

df['Score'] = ((df['DTI'] - min_val) / (max_val - min_val)) * 100
```
